## Renaming and combining the value of the categorical variables(s4aq6a,s4aq3,poverty)

```python
# Feature engineering
```

```python
autoyA['s4aq6a'].unique()
```

```python
pd.crosstab(autoyA['s4aq6a'], columns='Frequence')
```

```python
autoyA['Highest_educ_atte']= autoyA['s4aq6a']
autoyA.head(2)
```

```python
autoyA['Highest_educ_atte'].unique()
```

```python
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace('Missing', None)
```

```python
autoyA['Highest_educ_atte']
```

```python
pd.crosstab(autoyA['Highest_educ_atte'], columns='Frequences')
```

```python
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace('Don\'t Know', None)
```

```python
pd.crosstab(autoyA['Highest_educ_atte'], columns='Frequences')
```

```python
# Feature Engeneering
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'Not in class'},'No educ')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'Pre-primary'},'Nursery')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'Primary 1','Primary 2','Primary 3','Primary 4','Primary 5','Primary 6,7,8'},'Primary')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'Post primary 1','Post primary 2','Post primary 3','Post primary 4','Post primary 5','Post primary 6,7,8'},'Post primary')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'Secondary 1','Secondary 2','Secondary 3','Secondary 4','Secondary 5','Secondary 6'},'Secondary')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'University 1','University 2','University 3'},'Bachelor\'s degree')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'University 4','University 5'},'Master\'s degree')
autoyA['Highest_educ_atte'] = autoyA['Highest_educ_atte'].replace({'University 6'},'PHD/Professor')
```

```python
pd.crosstab(autoyA['Highest_educ_atte'], columns='Frequences')
```

```python
# Ordering the attributes of the variable
autoyA['Highest_educ_atte'] = pd.Categorical(autoyA.Highest_educ_atte, categories=['No educ','Nursery','Primary','Post primary','Secondary',"Bachelor's degree","Master's degree",'PHD/Professor'])
```

```python
pd.crosstab(autoyA['Highest_educ_atte'], columns='Frequences')
```
